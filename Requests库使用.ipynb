### 添加headers

爬虫时添加headers信息为了避免发生错误

```python
# 不加headers 爬取知乎
response = requests.get('https://www.zhihu.com/explore')

print(response.text) # 不能爬取
```

```python
# 添加heders 爬取知乎
headers = {
    'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'
}
response = requests.get('https://www.zhihu.com/explore',headers=headers)

print(response.text) # 成功爬取
```

### 上传cookies

```python
使用cookies参数，接收字典和RequestsCookieJar对象
```

#### 使用RequestsCookieJar对象

```python
from requests.cookies import RequestsCookieJar
from cookies import string # 本地导入一个cookie
import json

# 创建一个RequestsCookieJar对象
cookie_jar = RequestsCookieJar()
print(type(cookie_jar))

# 使用set方法向RequestsCookieJar对象中增加cookie信息
for i in json.loads(string):
    cookie_jar.set(i['name'],i['value'])

print(cookie_jar)

r = requests.session()
r.get('https://world.taobao.com/',cookies = cookie_jar)

# 没有登录淘宝账号无法查询'美食'相关信息，必须添加cookies进行模拟登录
product = r.get('https://s.taobao.com/search?q=%E7%BE%8E%E9%A3%9F&type=p&tmhkh5=&spm=a21wu.241046-us.a2227oh.d100&from=sea_1_searchbutton&catId=100',
               cookies=cookie_jar)

print(product.text)
```

#### 使用字典

```python
cookies_dict = dict()

for cookie in json.loads(string):
        cookies_dict[cookie['name']] = cookie['value']
s = requests.session()
r = s.get("https://world.taobao.com/", cookies=cookies_dict)
```

### 会话维持

会话对象
会话对象让你能够跨请求保持某些参数。
它也会在同一个 Session 实例发出的所有请求之间保持 cookie

```python
# 使用httpbin 测试网站，先set cookie,然后get cookie

requests.get('http://httpbin.org/cookies/set/number/123456789')
response = requests.get('http://httpbin.org/cookies')
print(response.text)  # 无法获取保存的cookies,因为这里相当于两次独立的访问
```

```python
# 使用Session方法

s = requests.Session()
s.get('http://httpbin.org/cookies/set/number/123456789')
response = s.get('http://httpbin.org/cookies')
print(response.text) 
```

### 超时设置

```python
response = requests.get('https://httpbin.org/get',timeout=0.1)
print(response.status_code)
```

```python
from requests.exceptions import ReadTimeout

try:
    response = requests.get('https://httpbin.org/get',timeout=0.1)
    print(response.status_code)
except ReadTimeout:
    print('timeout')
```
